```python
import pandas as pd
df = pd.read_excel('relatorio_semanal.xlsx')
display(df)
```

```python
faturamento = df['VALOR'].sum()
qtd_consultas = df['TIPO CONSULTA'].count()
ortopedista_nomes = df.loc[df['TIPO CONSULTA']=="Ortopedista","NOME PACIENTE"].to_string(index=False)
nutricionista_nomes = df.loc[df['TIPO CONSULTA']=="Nutricionista","NOME PACIENTE"].to_string(index=False)
psicologo_nomes = df.loc[df['TIPO CONSULTA']=="Psicólogo","NOME PACIENTE"].to_string(index=False)
dentista_nomes = df.loc[df['TIPO CONSULTA']=="Dentista","NOME PACIENTE"].to_string(index=False)
oftalmologista_nomes = df.loc[df['TIPO CONSULTA']=="Oftalmologista","NOME PACIENTE"].to_string(index=False)
valor_ortopedista = df.loc[df['TIPO CONSULTA']=="Ortopedista", "VALOR"].sum()
valor_nutricionista = df.loc[df['TIPO CONSULTA']=="Nutricionista", "VALOR"].sum()
valor_psicologo = df.loc[df['TIPO CONSULTA']=="Psicólogo", "VALOR"].sum()
valor_dentista = df.loc[df['TIPO CONSULTA']=="Dentista", "VALOR"].sum()
valor_oftalmologista = df.loc[df['TIPO CONSULTA']=="Oftalmologista", "VALOR"].sum()
qtd_cada_consulta = df['TIPO CONSULTA'].value_counts(sort=True, ascending=False)
print('O faturamento da semana foi de R$%.2f' %faturamento,'reais')
print('\nPacientes de ortopedia - Nomes:')
print(ortopedista_nomes)
print('Valor total de todas as consultas da especialidade: R$%.2f'%valor_ortopedista,'reais')
print('\n\nPacientes de nutrição - Nomes:')
print(nutricionista_nomes)
print('Valor total de todas as consultas da especialidade: R$%.2f'%valor_nutricionista,'reais')
print('\n\nPacientes de psicologia - Nomes:')
print(psicologo_nomes)
print('Valor total de todas as consultas da especialidade: R$%.2f'%valor_psicologo,'reais')
print('\n\nPacientes de dentista - Nomes:')
print(dentista_nomes)
print('Valor total de todas as consultas da especialidade: R$%.2f'%valor_dentista,'reais')
print('\n\nPacientes de oftalmologia - Nomes:')
print(oftalmologista_nomes)
print('Valor total de todas as consultas da especialidade: R$%.2f'%valor_oftalmologista,'reais')
print('\nA quantidade total de consultas na semana:',qtd_consultas,'consultas')
print('\nA quantidade de consultas em cada especialidade:')
print(qtd_cada_consulta)
    
```
